# 質的変数と量的変数
変数には「質的変数」と「量的変数」が存在する

*   質的変数：種類（属性）を区別するような変数⇒アンケートの満足度や血液型など。質的変数の中でも、男性・女性、喫煙習慣の有無といった２種類の値しかとらない質的変数を「2値変数」と呼ぶ
*   量的変数：量を表現する変数⇒テストの点数、身長など。


# 尺度水準
質的変数はさらに細かく４つに分類される

+ 名義尺度：単に分類するための変数（生徒番号、電話番号、性別など）
+ 順序尺度：順序関係や大小関係に意味のある変数（成績の順位、アンケートの満足度など）⇒大小関係にのみ意味があり、４位は８位より２倍出来がいい。という主張は成り立たない。
+ 間隔尺度：大小関係に加え、差に意味がある変数。（西暦、温度など）⇒30℃よりも60℃のほうが温度が2倍高いという主張が成り立つ。
+ 比例尺度：大小関係、差、比すべてに意味のある変数（長さや重さなど）⇒100cmは50cmの2倍ということに意味がある。

間隔尺度と比例尺度の区別を見分けるコツは「0が絶対的な無を表すか」⇒長さは0であれば長さがない。0℃は温度がないことを意味しない。



# 離散型変数と連続型変数
+ 離散型変数：とびとびの値をとり、隣り合う数値の間には値が存在しない（0,1,2・・）
+ 連続型変数：連続の値をとることができる変数。どんな2つの数字の間にも、必ず数字が存在する。（長さや重さ、時間など）

実際に扱うデータは連続変数であっても、測度精度が有限であるため、取りうる値はとびとびになってしまう。
例えば、身長のデータは170.3cmと170.4cmという値の間には数字が存在できないので、厳密には離散型変数に分類されてしまう。**測度精度の問題で離散型になってしまう変数は、多くの場合連続型変数として扱われる**

# まとめ
+ 生徒番号：名義尺度
+ 学年：離散型の間隔尺度
+ 握力：連続型の比例尺度
+ 上体起こし：離散型の比例尺度
+ 点数：連続型（微妙な判断）の比例尺度
+ 順位：離散型の順序尺度
